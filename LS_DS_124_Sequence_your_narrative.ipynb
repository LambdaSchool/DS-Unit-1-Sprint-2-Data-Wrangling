## Merge data

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

```python
print(income.shape)
print(lifespan.shape)
#this is using the default perameters
df = pd.merge(income, lifespan)
# This is doing the same thing as above but explicitly specifying the parameters
df = pd.merge(income, lifespan, on=['geo','time'], how='inner')
df.head()
```

```python
df = pd.merge(df, population)
print(df.shape)
df.head()
```

```python
entities['world_6region'].value_counts()
```

```python
entities['world_4region'].value_counts()
```

```python
entity_columns_to_keep = ['country', 'name', 'world_4region', 'world_6region']

entities = entities[entity_columns_to_keep]

entities.head()
```

```python
merged = pd.merge(df, entities, left_on='geo', right_on='country')

merged.head()
```

```python
merged = merged.drop('geo', axis='columns')
merged.head()
```

```python
merged = merged.rename(columns = {
    'country': 'country_code',
    'time': 'year', 
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income', 
    'life_expectancy_years': 'lifespan',
    'population_total': 'population',
    'name': 'country',
    'world_6region': '6region',
    'world_4region': '4region'
})
```

## Plot visualization

```python
import seaborn as sns
now = merged[merged.year == 2018]
now.head()
# sns.relplot(x=now['income'], y=now['lifespan'])
```

## Analyze outliers

```python
now.sort_values('income', ascending=False)
```

```python
now_qatar = now[now.country == 'Qatar']
now_qatar.head()
```

```python

sns.relplot(x='income', y='lifespan', hue='6region', size='population', sizes=(30,400), data=now)
plt.xscale('log')
plt.ylim(20,85)
plt.title("The World in 2018")
plt.text(x=now_qatar.income-5000, y= now_qatar.lifespan+1, s='Qatar')
plt.show()
```

```python
sns.relplot(x='income', y='lifespan', hue='6region', size='population', sizes=(30,500), data=now);
plt.xscale('log')
plt.ylim(20, 85)
plt.title("The World in 2018")
plt.text(x=now_qatar.income-5000, y=now_qatar.lifespan+1, s='Qatar')
plt.show()
```

## Plot multiple years

```python
years = [1818, 1918, 2018]

centuries = merged[merged.year.isin(years)]

fig = sns.relplot(x='income', y='lifespan', hue='6region', size='population', 
                  sizes=(30,400), col='year', data=centuries)
plt.text(x=now_qatar.income-5000, y= now_qatar.lifespan+1, s='Qatar')
plt.xscale('log')


# set titles on the individual "axes"
axes = fig.axes.flatten()
axes[0].set_title("1818")
axes[1].set_title("1918")
axes[2].set_title("2018")
plt.show()
```

## Point out a story

```python
years = [1918, 1938, 1978, 1998, 2018]

decades = merged[merged.year.isin(years)]
```

```python
sns.relplot(x='income', y='lifespan', hue='6region', size='population', col='year', data=decades)
plt.show()
```

```python
for year in years:
  sns.relplot(x='income', y='lifespan', hue='6region', size='population',
              data=merged[merged.year==year])
  plt.xscale('log')
  plt.xlim((150, 150000))
  plt.ylim((0, 90))
  plt.axvline(x=1000, color='grey')
  plt.axhline(y=40, color='grey')
```
