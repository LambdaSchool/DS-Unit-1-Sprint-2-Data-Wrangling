## Merge data

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

```python
# Merge income, lifespan and population dataframes
df = income.merge(lifespan).merge(population)

# See the merged df
print(df.head())
df.shape
```

```python
# Incorporate entities dataframe to add region and country name
entities.head()
```

```python
# Pull only needed attributes
entities[['country', 'name', 'world_6region']].head()
```

```python
# Merge entities with prior merged df; we only use country code as key 
# because country, name and region are static

df = pd.merge(df, entities[['country', 'name', 'world_6region']], left_on ='geo', right_on ='country')
```

```python
# Validate the merge
print(df.shape)
df.head()
```

```python
# Remove the geo and country features as they redundant
df = df.drop(columns=['geo', 'country'])

# Verify changes
df.head()
```

```python
# Rename columns/features
df = df.rename(columns={
    'time': 'year',
    'income_per_person_gdppercapita_ppp_inflation_adjusted':'income',
    'life_expectancy_years': 'lifespan',
    'population_total': 'population',
    'name': 'country',
    'world_6region': 'region'
})

df.tail()
```
