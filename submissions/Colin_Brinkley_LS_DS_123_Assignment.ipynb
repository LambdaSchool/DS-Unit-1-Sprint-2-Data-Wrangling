```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from IPython.display import display, Image
```

```python
example = Image(url='https://fivethirtyeight.com/wp-content/uploads/2017/09/mehtahickey-inconvenient-0830-1.png', width=400)
display(example)
```

```python
#Use a temporary dataset
fake = pd.Series([38, 3, 2, 1, 2, 4, 6, 5, 5, 33], index=range(1,11))

#Use the 538 styles
plt.style.use('fivethirtyeight')

#Define outer figure
fig = plt.figure()
fig.patch.set(facecolor='white')

#Define axis
ax = fake.plot.bar(color="#ED713A", width=0.9)
ax.set(facecolor='white')

#Title
ax.text(x=-1.8, y=44, s="'An Inconvenient Sequel: Truth To Power' is divisive", fontweight='bold', fontsize=12);

#Subtitle
ax.text(x=-1.8, y=41.5, s="IMDb ratings for the film as of Aug. 29", fontsize=11);

#X Label
ax.set_xlabel("Rating", fontweight='bold', fontsize=10)

#Y Label
ax.set_ylabel("Percent of total votes", fontweight='bold', fontsize=10)

#Set ticks and their labels
plt.xticks(np.arange(0, 10, 1), fontsize=9, color="0.7", rotation='horizontal')
plt.yticks(np.arange(0, 50, 10), ('0  ', '10  ', '20  ', '30  ', '40%'), fontsize=9, color="0.7");

#Removes frame
ax.set_frame_on(False)
```
