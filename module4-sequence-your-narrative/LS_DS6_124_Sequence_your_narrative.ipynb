## Merge data

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

```python
df = pd.merge(income, lifespan)
df.shape
```

```python
df = pd.merge(df, population)
df.shape
```

```python
entities['world_6region'].value_counts()
```

```python
entities['world_4region'].value_counts()
```

```python
entities_variables = ['country', 'name', 'world_6region']

entities = entities[entities_variables]

print(entities.shape)
entities.head()
```

```python
merged = pd.merge(df, entities, left_on='geo', right_on='country')
print(merged.shape)
merged.head()
```

```python
merged = merged.drop(['geo', 'country'], axis='columns')
```

```python
merged.head()
```

```python
merged = merged.rename(columns = {
    'time': 'year', 
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income', 
    'life_expectancy_years': 'lifespan',
    'population_total': 'population',
    'name': 'country',
    'world_6region': 'region'
})

merged.head()
```

## Plot visualization

```python
import seaborn as sns

now = merged[merged['year'] == 2018]

sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);
```

```python
qatar_now = now[(now.income > 80000) & (now.country == 'Qatar')].sort_values(by='income')
qatar_now
```

```python


sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);
plt.text(x=qatar_now.income-5000 , y=qatar_now.lifespan+1 , s='Qatar');
```

```python
years = [1818, 1918, 2018]

centuries = merged[merged.year.isin(years)]

sns.relplot(x='income', y='lifespan', hue='region', size='population', col='year', data=centuries);
plt.xscale('log');
```

```python
years = [1918, 1938, 1958, 1978, 1998, 2018]

decades = merged[merged.year.isin(years)]

sns.relplot(x='income', y='lifespan', hue='region', size='population', col='year', data=decades);
```

```python
for year in years:
    sns.relplot(x='income', y='lifespan', hue='region', size='population', 
                data=merged[merged.year==year])

    plt.xscale('log')
    plt.xlim((150, 150000))
    plt.ylim((0, 90))
    plt.title(year)
    plt.axvline(x=1000, color='grey')
```

```python
merged[(merged.year==1918) & (merged.lifespan > 50)]
```

```python
merged[(merged.year==2018) & (merged.lifespan < 50)]
```

```python
final_df = final_df.rename(columns = {
    'life_expectancy_years': 'lifespan',
    'population_total': 'population',
    'time': 'year',
    'name': 'country',
    'world_4region': 'region'
})
final_df.head()
```

```python
import seaborn as sns
```

```python
year2018 = final_df[final_df.year == 2018]
sns.relplot(x='income', y='lifespan', hue='region', size='population', data=year2018);
```

```python
high_country = year2018[year2018['income'] > 100000].sort_values(by='income')
high_country
```

```python
df_generation = final_df[final_df['year'].isin(generations)]
sns.relplot(x='income', y='population', hue='region', size='population', data=df_generation, col='year', sizes=(20, 200), s=df_generation['population']**2)
plt.xscale('linear')
```

```python
#check marker size changed from default
sns.relplot(x='income', y='population', hue='region', size='population', data=df_generation, col='year')
plt.xscale('symlog')
```

```python
for year in generations:
  sns.relplot(x='income', y='lifespan', hue='region', size='population', 
             data=final_df[final_df.year==year])
  plt.xscale('log')
  plt.xlim((150,150000))
  plt.ylim((0, 100))
  plt.title(year)
  plt.axvline(x=5000, color='black')
  plt.axhline(y=50, color='black')
```

```python
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"];
sns.palplot(sns.xkcd_palette(colors));

from google.colab import widgets
tb=widgets.TabBar([str(year) for year in generations]);
for tab, year in zip(tb, generations):
  sns.relplot(x='income', y='lifespan', hue='region', size='population',
             sizes=(2, 200), data=final_df[final_df['year']==year])
  plt.xscale('log')
  plt.xlim((150, 150000))
  plt.ylim((0, 100))
  plt.title(year)
  plt.xlabel(xlabel='Income', fontsize=9, fontweight='bold')
  plt.ylabel(ylabel='Lifespan', fontsize=9, fontweight='bold')
  plt.text(x=2500, y=95, s='Gapminder Data', fontsize=8, alpha=0.75)



```
