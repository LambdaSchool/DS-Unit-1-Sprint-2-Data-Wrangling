## Merge data

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

```python
df = pd.merge(income,lifespan)
df.shape
#******** MERGE taking place HEERRREEE********
```

```python
df = pd.merge(df, population)
df.shape
#******** MERGE taking place HEERRREEE********
```

```python
entities['world_6region'].value_counts() #we are only grabbing specific varibles from our entities columns
```

```python
entities['world_4region'].value_counts()
```

```python
entities_variables = ['country', 'name', 'world_6region']

entities = entities[entities_variables] #what we are doing here is putting our new entities_variable inside of out other variable entities "Merging" them

#******************* this is where the merging is taking place^^^^^^^^^^^^^^^^^^^^^^^^
```

```python
print(entities.shape)
entities.head()
#*********** PROOOF THE THE MERGE HAS TAKEN PLACE************
```

```python
merge = pd.merge(df, entities, left_on='geo', right_on='country')
print(merge.shape)
merge.head()
```

```python
merge = merge.drop(['geo', 'country'], axis='columns')
```

```python
merge.head()
#*****geo and country have now been droped from the dataframe
```

```python
#******** NOW WE ARE RENAMING the columns********
merge = merge.rename(columns= { 'time': 'year', 
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income', 
    'life_expectancy_years': 'lifespan',
    'population_total': 'population',
    'name': 'country',
    'world_6region': 'region'})

# NOTICE that a ':' is use to rename the columns VERY IMPORTANT
```

```python
merge.head()
```

## Plot visualization

```python
import seaborn as sns
now = merge[merge['year']== 2018]

sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);
```

```python
qatar_now = now[(now.income > 80000) & (now.country == 'Qatar')].sort_values(by='income')
qatar_now
```

```python
sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);
plt.text(x=qatar_now.income-5000 , y=qatar_now.lifespan+1 , s='Qatar'); # plt.text is for the over lay labeling the the qatar point on the chart
```

```python
years = [1818, 1918, 2018]

centuries = merge[merge.year.isin(years)]
sns.relplot(x='income', y='lifespan', hue='region', size='population', col='year', data=centuries);
plt.xscale('log');
```

```python
years = [1918, 1938, 1958, 1978, 1998, 2018]

decades = merge[merge.year.isin(years)]

sns.relplot(x='income', y='lifespan', hue='region', size='population', col='year', data=decades);
```

```python
for year in years:
    sns.relplot(x='income', y='lifespan', hue='region', size='population', 
                data=merge[merge.year==year])

    plt.xscale('log')
    plt.xlim((150, 150000))
    plt.ylim((0, 90))
    plt.title(year)
    plt.axhline(y=30, color='grey')
```

```python
merge[(merge.year==1918) & (merge.lifespan > 50)]
```

```python
merge[(merge.year==2018) & (merge.lifespan < 50)]
```
