## Work with strings

For machine learning, we usually want to replace strings with numbers

We can get info about which columns have a datatype of "object" (strings)

```
df.select_dtypes('object').info()
```

### Convert `int_rate`

Define a function to remove percent signs from strings and convert to floats

```
string='17.97%'

def remove_percent(string):
    return float(string.strip('%'))

remove_percent(string)
```

Apply the function to the `int_rate` column

```
df['int_rate'] = df['int_rate'].apply(remove_percent)
```

```
df['int_rate'].head()
```

### Clean `emp_title`

Look at top 20 titles

```
df['emp_title'].value_counts().head(20)
```

How often is `emp_title` null?

```
df['emp_title'].isnull().sum()
```

Clean the title and handle missing values

```
examples = ['owner', 'Supervisor ', 
            ' Project manager', np.nan]

def clean_title(x):
    if isinstance(x, str):        
        return x.strip().title()
    else:
        return 'Unknown'

for example in examples:
    print(clean_title(example))
```

```
df['emp_title'] = df['emp_title'].apply(clean_title)
```

```
df['emp_title'].value_counts().head(20)
```

### Create `emp_title_manager`

pandas documentation: [`str.contains`](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.str.contains.html)

```
# Boolean column, True if `emp_title` contains 'Manager',
# otherwise False
```

```
df['emp_title_manager'] = df['emp_title'].str.contains('Manager')
```

```
df['emp_title_manager'].value_counts()
```

- Convert the `term` column from string to integer.

- Make a column named `loan_status_is_great`. It should contain the integer 1 if `loan_status` is "Current" or "Fully Paid." Else it should contain the integer 0.

- Make `last_pymnt_d_month` and `last_pymnt_d_year` columns.



```
# 1: Replicate the lesson code.

# import statements
import pandas as pd
import numpy as np

# set options
pd.options.display.max_columns = 200
pd.options.display.max_rows = 200

# function definitions
# define Data sanity check functions
def all_numeric(df):
    return all((df.dtypes==np.number) | 
               (df.dtypes==bool))

def no_nulls(df):
    return not any(df.isnull().sum())

def ready_for_sklearn(df):
    return all_numeric(df) and no_nulls(df)
  
# define fcn to remove % from a string and return a float of the resultant number
def remove_percent(string):
  return float(string.strip('%'))

# define fcn to clean employee title field
def clean_title(input):
    if isinstance(input, str):
        return input.strip().title()
    return "Unknown"

# Get .zip file
!wget https://resources.lendingclub.com/LoanStats_2018Q3.csv.zip

# unzip same
!unzip LoanStats_2018Q3.csv.zip

# shell cmd to look at first ten rows
!echo
!echo first ten rows of data
!head LoanStats_2018Q3.csv

# and last ten
!echo
!echo last ten rows of data
!tail LoanStats_2018Q3.csv
!echo

# load data, skipping first and last two rows
df = pd.read_csv('LoanStats_2018Q3.csv', skiprows=1, skipfooter=2)

# print shape of data
print("\nData Frame shape:", df.shape)

# print data
print("\nRandom 12 rows of Data\n:", df.sample(12))

# print data transposed
print("\nFirst five rows of Data transposed\n:", df.head().T)

# check data sanity
print("\n Data Sanity checks:")
print("Data is all numeric", all_numeric(df))
print("Data has no nulls", no_nulls(df))
print("Data is ready for sklearn", ready_for_sklearn(df))

# print list of object types
print("\n Data object types:")
print(df.select_dtypes('object').info())

# remove percents from interest rate column and convery to number
df['int_rate'] = df['int_rate'].apply(remove_percent)
  
# clean employee title
df['emp_title'] = df['emp_title'].apply(clean_title)

# Create emp_title_manager column
df['emp_title_manager'] = df['emp_title'].str.contains('Manager')

# print a counts of same
print("\n Count of managers:")
print(df['emp_title_manager'].value_counts())

# convert issue date to datetime format
df['issue_d'] = pd.to_datetime(df['issue_d'], 
                               infer_datetime_format=True)

# create issue year, month columns
df['issue_year']  = df['issue_d'].dt.year
df['issue_month'] = df['issue_d'].dt.month

# print shape of data
print("\nData Frame shape:", df.shape)
```

```
# 2: Convert the term column from string to integer
#df['term'].isna().sum()
def convert_term(input):
    return int(input.strip(' months'))

#convert_term('36 months')
df['term'] = df['term'].apply(convert_term)
```

```
# 3: Make a column named loan_status_is_great. Contains the integer 1 if loan_status is "Current" or "Fully Paid." Else it should contain the integer 0.
def loan_status(input):
  if input == "Current" or input == "Fully Paid.":
    return 1
  return 0

#loan_status('Bum')
df['loan_status_is_great'] = df['loan_status'].apply(loan_status)
#df['loan_status_is_great'].value_counts()
```

```
# 4: Make last_pymnt_d_month and last_pymnt_d_year columns
pd.set_option('precision',0)
last_pymnt_d = pd.to_datetime(df['last_pymnt_d'], 
                                  infer_datetime_format=True)
df['last_pymnt_d_year']  = last_pymnt_d.dt.year
df['last_pymnt_d_month'] = last_pymnt_d.dt.month
```

```
# print shape of data
print("\nData Frame shape:", df.shape)
```
