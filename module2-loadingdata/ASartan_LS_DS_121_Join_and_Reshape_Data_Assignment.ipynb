```python
import pandas as pd
```

```python
#Figuring out which dataframes and columns we need
products = pd.read_csv('products.csv')
products.head()
products.tail()
```

```python
order_products_prior = pd.read_csv('order_products__prior.csv')
order_products_prior.head()
```

```python
order_products_train = pd.read_csv('order_products__train.csv')
order_products_train.head()
```

```python
#Merging order_products dataset

order_products = pd.concat([order_products_prior, order_products_train])
order_products.shape
order_products.tail()
```

```python
#Checking if merging was done correctly

assert len(order_products_prior) + len(order_products_train) == len(order_products)
```

```python
#Merging dataframes: order_products with product_name. All we need is jst two columns: product_id and product_name

merged = products[['product_id', 'product_name']].merge(order_products[['product_id']])
merged.sample(50)
```

```python
#Getting the counts of the top 10 most frequently ordered products

merged['product_name'].value_counts(dropna=False).head(10)
```
