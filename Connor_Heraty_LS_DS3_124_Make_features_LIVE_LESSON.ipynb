## Stretch Goal - Playground

First I will create a subset of features that will reveal relevant insights about our data.

```python
sub_df = loan_df[['loan_amnt', 'int_rate','grade','sub_grade','emp_title','home_ownership','emp_length','annual_inc','loan_status','addr_state','avg_cur_bal','pub_rec_bankruptcies']]
```

```python
sub_df.head()
```

```python
sub_df.isnull().sum()
```

```python
sub_df = sub_df.dropna()
```

```python
loan_amnt = (sub_df
            .groupby(['emp_title'])
            .loan_amnt.agg(['mean'])
            .rename(columns={'mean':'Average Loan Amount'})
            .reset_index())
```

```python
annual_income = (sub_df
            .groupby(['emp_title'])
            .annual_inc.agg(['mean'])
            .rename(columns={'mean':'Average Annual Income','emp_title':'Title'})
            .reset_index())
```

```python
loan_to_income = pd.merge(loan_amnt, annual_income, on='emp_title')
loan_to_income.head()
```

The goal of this chart is to plot the average loan amount and average annual income for each employee group.
- Operation managers appear to have a significantly higher average loan to income ratio.

```python
import altair as alt

alt.Chart(loan_to_income,height=400,width=700).mark_circle(size=100).encode(
    x='Average Annual Income',
    y='Average Loan Amount',
    tooltip='emp_title',
    color='emp_title',
    
    
).interactive()
```

```python
invest_grade = (sub_df
            .groupby(['grade','sub_grade'])
            .loan_amnt.agg(['mean','count','sum'])
            .rename(columns={'mean':'Average Loan by Grade', 'count':'Total Loans by Grade', 'sum':'Total Amount of Loans by Grade'})
            .reset_index())
```

```python
invest_grade.head()
```

The plot is designed to show the distribution of loans originated based on the investment grade of the borrower. 
- The colors represent the distribution of loan origination per borrower subgrade (1-5) within the primary grade (A-G).
- There is a significant drop off in loan origination after the 'C' borrower grade.

```python
alt.Chart(invest_grade,height=400,width=700).mark_bar().encode(
    x='grade',
    y='Total Amount of Loans by Grade',
    color='sub_grade'
)
```

```python
bankruptcies = (sub_df
            .groupby(['grade','pub_rec_bankruptcies'])
            .loan_amnt.agg(['mean','sum'])
            .rename(columns={'mean':'Average Loan by Grade', 'count':'Total Loans by Grade', 'sum':'Total Amount of Loans by Grade'})
            .reset_index())
```

This graphs the total amount of loans originated by borrower bankruptcy and investment grade.
- The majority of loan originations were to individuals with no public bankruptcies.
- There is a significant drop off with each succeeding bankruptcy.
- The second graph (using the average loan amount) shows that when loans were originated for borrowers with multiple bankruptcies, the average amount was similar to the average. 

```python
alt.Chart(bankruptcies).mark_bar().encode(
    x='grade',
    y='Total Amount of Loans by Grade',
    color='grade',
    column='pub_rec_bankruptcies'
)
```

```python
alt.Chart(bankruptcies).mark_bar().encode(
    x='grade',
    y='Average Loan by Grade',
    color='grade',
    column='pub_rec_bankruptcies'
)
```
