#Reproduce FiveThirtyEight Plot

```python
from IPython.display import display, Image

url = 'https://fivethirtyeight.com/wp-content/uploads/2019/07/choi-foul-0625-2-1.png?w=1150'
example_pic = Image(url=url, width=400)

display(example_pic)
```

```python
#importing the raw data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/foul-balls/foul-balls.csv')
df.head()
```

```python
df.plot(x='predicted_zone', y='exit_velocity', kind='barh');
```

```python
df = df.sort_values(by=['predicted_zone'])
df.plot(x='predicted_zone', y='exit_velocity', kind='barh');
```

```python
zone1_group = df.groupby('predicted_zone').get_group(1)
zone2_group = df.groupby('predicted_zone').get_group(2)
zone3_group = df.groupby('predicted_zone').get_group(3)
zone4_group = df.groupby('predicted_zone').get_group(4)
zone5_group = df.groupby('predicted_zone').get_group(5)
zone6_group = df.groupby('predicted_zone').get_group(6)
zone7_group = df.groupby('predicted_zone').get_group(7)
```

```python
zone1_group.plot(x='predicted_zone', y='exit_velocity', kind='barh');
```

```python
zone1 = zone1_group['predicted_zone'].value_counts()
zone2 = zone2_group['predicted_zone'].value_counts()
zone3 = zone3_group['predicted_zone'].value_counts()
zone4 = zone4_group['predicted_zone'].value_counts()
zone5 = zone5_group['predicted_zone'].value_counts()
zone6 = zone6_group['predicted_zone'].value_counts()
zone7 = zone7_group['predicted_zone'].value_counts()
```

```python
df2 = pd.Series([zone1, zone2, zone3, zone4, zone5, zone6, zone7]);
df2
```

```python
#cleaning this up
#making the dataframe for number of times each foul ball goes into each zone
df2 = pd.Series([5,6,223,208,83,104,277], [7,6,5,4,3,2,'Zone 1'])
df3 = pd.Series([5,6,120,125,40,50,150],[7,6,5,4,3,2,'Zone 1'] )
plt.style.use('fivethirtyeight');

plt.figure(facecolor='white', figsize=(7, 7))

df2.plot.barh(color='#a7a7a7',width=0.7)
df3.plot.barh(color='#b1e261',width=0.7)


plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True

plt.title('The hardest hit fouls seem to land in unprotected areas',
          fontsize=10,
          loc='left',
          x=0,
          y=1.2,
          fontweight='bold')
plt.text(x=0, y=7.4, s='Foul balls by the stadium zone they landed in and their exit velocity, among \n 906 fouls hit this season in the most foul heavy day at the 10 MLB \nstadiums that produced the most fouls as of June 5', fontsize=10)

plt.yticks(rotation=0, color='black');
```
