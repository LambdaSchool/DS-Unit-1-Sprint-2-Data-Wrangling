```python
#Import required libraries

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load required data in pandas dataframe
income = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--income_per_person_gdppercapita_ppp_inflation_adjusted--by--geo--time.csv')
```

```python
lifespan = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--life_expectancy_years--by--geo--time.csv')
```

```python
population = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--datapoints--population_total--by--geo--time.csv')
```

```python
entities = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--entities--geo--country.csv')
```

```python
concepts = pd.read_csv('https://raw.githubusercontent.com/open-numbers/ddf--gapminder--systema_globalis/master/ddf--concepts.csv')
```

```python
# find shape of the dataframes
income.shape, lifespan.shape, population.shape, entities.shape, concepts.shape
```

```python
lifespan.head()
```

```python
population.head()
```

```python
entities.head()
```

```python
concepts.head()
```

```python
#merge income and lifespan dataframes with geophraphic location and time (years) as indexes and inner join.

# Inner join = An inner join combines two DataFrames based on a join key and returns a new DataFrame that contains only those rows that have matching values in both of the original DataFrames.

#Inner join yields a DataFrame that contains only rows where the value being joins exists in BOTH tables


df = pd.merge(income, lifespan, on=['geo', 'time'], how='inner')


```

```python
df.head()
```

```python
df.shape
```

```python
#merging above dataframe df with population dataframe

df2 = pd.merge(df, population)
df2.head()
```

```python
entities.head()
```

```python
subset_cols = ['country', 'name', 'world_4region', 'world_6region']
merged = pd.merge(df, entities[subset_cols], left_on='geo', right_on='country')
merged = merged.drop(columns=['geo'])
merged.head()
```

```python
mapping_1 = {
    'time': 'year',
    'country': 'country_code',
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income',
    'population_total': 'population',
    'world_4region': '4region',
    'world_6region': '6region',
    'life_expectancy_years': 'lifespan'
}
mapping_2 = {'name': 'country'}
merged = merged.rename(columns=mapping_1)
merged = merged.rename(columns=mapping_2)
merged.head()
```

```python
merged.dtypes
```

```python
merged.describe()
```

```python
merged.describe(exclude='number')
```

```python
merged['country'].value_counts()
```

```python
mask_usa = merged['country'] == 'United States'
usa = merged[mask_usa]
usa.head()
```

```python
usa[usa['year'].isin([1818, 1918, 2018])]
```

```python
mask_china = merged['country'] == 'China'
china = merged[mask_china]
china[china['year'].isin([1818, 1918, 2018])]
```

```python
# Data Visualization

import seaborn as sns

#now = merged[merged['year'].isin([2018])]
now = merged[merged['year'] == 2018]
now.head()
```

```python
sns.relplot(x='income', 
            y='lifespan', 
            hue='6region',
            data=now);
plt.xscale('log')
plt.ylim([30, 90])
plt.title('The World in 2018');
```

```python
#Analyze outliers

now.sort_values('income')
```

```python
now.sort_values('income', ascending=False)
```

```python
now.sort_values('lifespan', ascending=False)
```

```python
# Using numpy to find correlation coeeficient between lifespan and income

np.corrcoef(now['lifespan'], now['income'])
```

```python
df_1990 = merged[merged['year'] == 1990]
np.corrcoef(df_1990['lifespan'], df_1990['income'])[0][1]
```

```python
years = range(1900, 2019)
corr_list = []

for year in years:
  df_year = merged[merged['year'] == year]
  corr = np.corrcoef(df_year['lifespan'], df_year['income'])[0][1]
  corr_list.append(corr)
  
corr_arr = np.array(corr_list)
```

```python
year_arr = np.array(years)
```

```python
np.argmax(corr_arr)
```

```python
year_arr[34]
```

```python
plt.scatter(year_arr, corr_arr);
```

```python
now_qatar = now[now['country'] == 'Qatar']
now_qatar.head()
```

```python
now_qatar['income'].values[0]
```

```python
sns.relplot(x='income', 
            y='lifespan', 
            hue='6region', 
            data=now
           )
plt.xscale('log')
plt.ylim([30, 90])
plt.title('The World in 2018')
plt.text(x=now_qatar['income'].values[0]-2000, 
         y=now_qatar['lifespan'].values[0]-2, 
         s='Qatar')
```

```python
sns.relplot(x='income', 
            y='lifespan', 
            hue='6region', 
            data=now
           )
plt.xscale('log')
plt.ylim([30, 90])
plt.title('The World in 2018')
plt.text(x=now_qatar['income'].values[0]-10000, 
         y=now_qatar['lifespan'].values[0]-7, 
         s='Qatar')
plt.arrow(x=now_qatar['income'].values[0],
         y=now_qatar['lifespan'].values[0],
         dx=0,
         dy=-5
         )
```

```python
# Plot multiple years

years = [1918, 1948, 1978, 2008]

for year in years:
  df_year = merged[merged['year']==year]
  sns.relplot(x='income', 
              y='lifespan', 
              hue='6region',
              data=df_year);
  plt.xscale('log')
  plt.ylim([0, 90])
  plt.title(f'The World in {year}')
```

```python
# Using Plotly

import plotly.express as px

px.scatter(merged, 
           x='income', 
           y='lifespan',
           animation_frame='year',
           color='6region',
           log_x=True,
           hover_name='country'
          )
```
