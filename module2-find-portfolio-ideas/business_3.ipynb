```
df_ba_us = pd.read_excel('/content/gdrive/My Drive/BA_US.xlsx') # sep='\t')
```

```
def strip_pull(column):
  now_stripped = []
  for row in column:
    now_stripped.append(row.split(' ')) # ','
  return now_stripped
```

```
time_col = strip_pull(df_ba_us['Time'])
```

```
def separating_two(abc):
  x1 = []
  y2 = []
  for sub in abc:
    x1.append(sub[0])
    y2.append(sub[1])
  return x1, y2
```

```
x1, y2 = separating_two(time_col)
test_list = list(map(int, x1))
```

```
year_us = pd.DataFrame(test_list)
```

```
year_us.head()
```

```
us_formation = df_ba_us['US'].values
```

```
year_us['US'] = us_formation
```

```
year_us.head(15)
```

```
year_us = year_us.rename(columns={0: 'Years'})
```

```
year_us.duplicated(subset='Years').value_counts()
```

```
years_average = year_us.groupby('Years', as_index=False)['US'].mean()
```
