###Reproduce with real data

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
df.head()
```

```python
df.dtypes
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.describe()
```

```python
df['timestamp'].min()
```

```python
df['timestamp'].max()
```

```python
df = df.set_index('timestamp')

df.head()
```

```python
df['2017-08-29']
```

```python
lastday = df['2017-08-29']
lastday_filtered = lastday[lastday['category'] == 'IMDb users']

lastday_filtered.head()
```

```python
lastday_filtered['respondents'].plot()
```

```python
lastday_filtered['category'].value_counts()
```

```python
pct_columns = [f'{i}_pct' for i in range(1,11)]
pct_columns
```

```python
final = lastday_filtered.tail(1)
```

```python
final
```

```python
final[pct_columns].T
```

```python
plot_data = final[pct_columns].T
plot_data.index = range(1,11)
```

```python
plot_data
```

```python
plt.style.use('fivethirtyeight')

ax = plot_data.plot.bar(color='#ed713a', width=0.9)

plt.xlabel('Rating', fontsize=9, fontweight='bold')
plt.ylabel('Percent of total votes', fontsize=9, fontweight='bold')
plt.title('`An Inconvenient Sequel: Truth to Power` is divisive',
         fontsize=12,
         x=-0.1, y=1.1, loc='left',
         fontweight='bold',
         fontname='Tahoma')
plt.text(x=-1.7, y=plot_data.max() + 4, s='IMDb ratings for the film as of Aug. 29',
        fontsize=11)
plt.xticks(rotation=0, color='#a7a7a7', fontsize=8, clip_on='false')
plt.yticks(range(0,50,10), labels=[f'{i}' if i != 40 else f'{i}%' for i in range(0,50,10)],
          color='#a7a7a7', fontsize=8, clip_on='false')
ax.set(facecolor='white')
fig = plt.figure(facecolor='white', edgecolor='white')
fig.patch.set_facecolor('white')

legend = ax.legend()
legend.remove()

display(fig)
```
