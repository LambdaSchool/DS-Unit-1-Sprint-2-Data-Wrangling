## Join Data Practice

These are the top 10 most frequently ordered products. How many times was each ordered? 

1. Banana
2. Bag of Organic Bananas
3. Organic Strawberries
4. Organic Baby Spinach 
5. Organic Hass Avocado
6. Organic Avocado
7. Large Lemon 
8. Strawberries
9. Limes 
10. Organic Whole Milk

First, write down which columns you need and which dataframes have them.

Next, merge these into a single dataframe.

Then, use pandas functions from the previous lesson to get the counts of the top 10 most frequently ordered products.

```python

#We will need the product name and the product id in products.csv, we will also need a count of the frequency of each product id in order products train and order products prior sets

#import the important libraries
import pandas as pd
from collections import Counter
```

```python
#get the aisles dataframe
aisles = pd.read_csv('aisles.csv')
print(aisles.shape)
aisles.head()
```

```python
#get the departments dataframe
departments = pd.read_csv('aisles.csv')
print(departments.shape)
departments.head()
```

```python
#get the order products prior dataframe
order_products__prior = pd.read_csv('order_products__prior.csv')
print(order_products__prior.shape)
order_products__prior.head()
```

```python
#get the order_products__train dataframe
order_products__train = pd.read_csv('order_products__train.csv')
print(order_products__train.shape)
order_products__train.head()
```

```python
#get the order dataframe
orders = pd.read_csv('orders.csv')
print(orders.shape)
orders.head()
```

```python
#get the products dataframe
products = pd.read_csv('products.csv')
print(products.shape)
products.head()
```

```python
#Combine the order products prior and the order products train
order_products = pd.concat([order_products__prior, order_products__train])
print(order_products.shape)
order_products.head()
```

```python
#check to make sure the files properly merged
assert (order_products__prior.shape[0] + order_products__train.shape[0]) == order_products.shape[0]
```

```python
columns = ['order_id', 'product_id']
#merge the product ids and product names
merged = pd.merge(orders, order_products[columns], how='inner', on='order_id')

merged
```

```python
final = pd.merge(merged, products[['product_id', 'product_name']], how='inner', on='product_id')
final
```

```python
#Get the counts for the top 10 most frequent ordered products
# most_ordered = Counter(final['product_name'])
# print most_ordered.most_common(10)

df = pd.DataFrame(final, columns = ['product_id', 'order_id', 'product_name'])
df
```

```python
dfpivot = df.pivot_table(index=['product_name'], aggfunc = 'count')
```

```python
dfpivot.sort_values(by=['product_id'], ascending = False)
```
