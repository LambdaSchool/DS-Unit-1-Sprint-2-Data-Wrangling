```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')
col=df.columns.to_list()
col=col[-10:] 
fake=df.loc[80052, col]
```

```python
# Your Work Here
fig=plt.figure()
ax=fake.plot.bar(color='C1', width = .9, zorder=3)

#copy the stuff from class
ax.patch.set_facecolor('white')
fig.patch.set_facecolor('white')
plt.title("'An Inconvient Sequel: Truth to Power' is devisive", fontsize=12, fontweight='bold', y=1.1, x=-.1, loc='left')
ax.text(-1.7,42, 'IMDb ratings for the film as of Aug. 29', fontsize=12)
ax.set_xticklabels(range(1,11), rotation=0, color='grey')
ax.set_yticklabels(range(0,60,10), color='grey')
ax.set_yticks(range(0,50,10))
plt.xlabel('Rating',fontsize=10, fontweight='bold', color='black')
plt.ylabel('Percent of total votes',fontsize=10, fontweight='bold', color='black')
#remove the outer bars
ax.spines['left'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

#show the grid
plt.grid(True)
ax.grid(zorder=0, color='grey')

#add the text at the bottom
plt.text(-1.7,-8.5, 'FiveThirtyEight', color='grey', fontsize='8')
plt.text(8 ,-8.5, 'SOURCE: IMDb', color='grey', fontsize='8')
plt.text(-1.8,-6, '___________________________________________________________________________', color='grey')

plt.show()
```

```python
#import the data
candy_url='https://github.com/fivethirtyeight/data/raw/master/candy-power-ranking/candy-data.csv'
candyDF=pd.read_csv(candy_url)
#eliminate the columns we don't need
candyCol=candyDF.columns.to_list()
candyCol=(candyCol[1:-1])
candyDF=candyDF.drop(candyCol, axis=1)
#sort the data so its in the right order
candyDF=candyDF.sort_values(by='winpercent', ascending=True)
```

```python
fig=plt.figure(figsize=[6.4,17])
ax=fig.add_subplot(111)
ax.barh(candyDF['competitorname'],candyDF['winpercent'])
```
