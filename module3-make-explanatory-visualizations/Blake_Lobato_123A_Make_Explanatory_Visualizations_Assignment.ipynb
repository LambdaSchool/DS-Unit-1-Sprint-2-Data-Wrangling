```python
from IPython.display import display, Image

url2 = 'https://fivethirtyeight.com/wp-content/uploads/2014/04/silver-age-congress-1.png?w=1150'
example2 = Image(url=url2, width=400)

display(example2)
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/congress-age/congress-terms.csv')
df.head(5)
```

```python
df.dtypes
```

```python
cut_df = df.drop(columns=['congress','chamber','bioguide','middlename', 'suffix'])
cut_df.head()
```

```python
display(example2)
```

```python
dems = cut_df[cut_df['party'] == 'D']
dems
```

```python
rep = cut_df[cut_df['party'] == 'R']
rep
```

```python
cut_df['termstart'] = pd.to_datetime(cut_df['termstart'],infer_datetime_format=True)
cut_df['termstart']
```

```python
dems['termstart'] = pd.to_datetime(dems['termstart'],infer_datetime_format=True)
dems['termstart']
rep['termstart'] = pd.to_datetime(rep['termstart'],infer_datetime_format=True)
rep['termstart']
```

```python
dems['year'] = dems['termstart'].dt.year
dems['year']
```

```python
rep['year'] = rep['termstart'].dt.year
rep['year']
```

```python
dems.head()
```

```python
average_age_at_term_D = dems.groupby(['year']).mean()
average_age_at_term_R= rep.groupby(['year']).mean()
```
